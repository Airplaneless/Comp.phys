```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from random import choice
from math import sqrt
from tqdm import tqdm
```

```python
def grow(A, bounds):    
    def update(p):
        A[p[0]][p[1]] = 1.0
        for m in range(-1, 2):
            for n in range(-1, 2):
                if not set([p[0], p[1]]) & set([0, A.shape[0]-1, A.shape[1]-1]) and not A[p[0]+m][p[1]+n]:
                    bounds.append([p[0]+m, p[1]+n])
        bounds.remove([p[0], p[1]])

    point = choice(bounds)
    update(point)
    return A, r, mass
```

### Модель роста Эдена на решетке 60х60 при 5000 итерациях

```python
A = np.zeros((60, 60))
bounds = list([[60/2, 60/2]])
for i in tqdm(range(5000)):
    A = grow(A, bounds)
    if i%100 == 0:
        print '{} iteration'.format(i)
        plt.imshow(A)
        plt.show()
```

```python
def mass(A, pc, radius):
    m = A[pc[0]][pc[1]]
    for i in xrange(len(A)):
        for j in xrange(len(A[i])):
            p = np.array([i, j])
            r = np.linalg.norm(p - pc)
            if r<radius:
                m+=A[p[0]][p[1]]
    return m
```

## Оценка фрактальной размерности 

```python
x = np.arange(1, 20)
y = [mass(A, np.array([30, 30]), r) for r in x]

regr = LinearRegression()
regr.fit(np.log(x).reshape(-1,1), np.log(y))
print 'df = {}'.format(regr.coef_)

plt.plot(np.log(x), np.log(y), 'go')
plt.plot(np.log(x), regr.predict(np.log(x).reshape(-1,1)), 'r')
plt.xlabel('lg(r)')
plt.ylabel('lg(M)')
plt.title(u"График зависимости lnM от lnr")
plt.show()
```
