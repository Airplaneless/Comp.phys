## Генерация ячеечной перколяции для различных вероятностей 

### Для квадратной решетки 4x4 

```python
percolate.init = True
for p in np.arange(0, 1, 0.025):
    print 'p = {}'.format(p)
    A = percolate(4,4,p)
    plt.imshow(A, cmap=plt.cm.jet)
    plt.show()
```
