```python
dataX0, dataY0 = analit_solve(T_0, T_INIT, GAMMA, 6, 60*60)

dt1 = 6
dataX, dataY = solve(T_0, T_INIT, GAMMA, dt1, 60*60)
delta1 = dataY0[int(60/6)] - dataY[int(60/dt1)]

dt2 = 3
dataX1, dataY1 = solve(T_0, T_INIT, GAMMA, dt2, 60*60)
delta2 = dataY0[int(60/6)] - dataY1[int(60/dt2)]

dt3 = 1.5
dataX2, dataY2 = solve(T_0, T_INIT, GAMMA, dt3, 60*60)
delta3 = dataY0[int(60/6)] - dataY2[int(60/dt3)]

dt4 = 0.6
dataX3, dataY3 = solve(T_0, T_INIT, GAMMA, dt4, 60*60)
delta4 = dataY0[int(60/6)] - dataY3[int(60/dt4)]

dt5 = 0.3
dataX4, dataY4 = solve(T_0, T_INIT, GAMMA, dt5, 60*60)
delta5 = dataY0[int(60/6)] - dataY4[int(60/dt5)]

plt.plot(dataX, dataY, 'r', label='dt=6s')
plt.plot(dataX1, dataY1, 'b', label='dt=3s')
plt.plot(dataX2, dataY2, 'g', label='dt=1.5s')
plt.plot(dataX3, dataY3, 'y', label='dt=0.6s')
plt.plot(dataX4, dataY4, '--', label='dt=0.3s')

plt.legend(loc='upper right')
plt.xlabel("Time, s")
plt.ylabel("Temp, Celsius")
plt.savefig('compare.PNG')
```

```python
delta_ls = [delta1, delta2, delta3, delta4, delta5]
dt_ls = [dt5, dt4, dt3, dt2, dt1]
plt.plot(dt_ls, delta_ls)
plt.xlabel('Time step, s')
plt.ylabel('Abs error')
plt.savefig('error.PNG')
```
