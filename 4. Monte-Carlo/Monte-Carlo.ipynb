```python
from random import random
from math import sqrt
from math import pi
from math import log
import numpy as np
import matplotlib.pylab as plt
from tqdm import tqdm
```

```python
def guess(n, under_func):
    nc = 0.0
    for i in range(n):
        y, x = random(), random()
        if under_func(x, y):
            nc+=1
    return nc/n
```

```python
def guess2(n, func):
    integ = 0.0
    for i in range(n):
        x = random()
        integ += func(x)
    return integ/n
```

```python
def func(x):
    return sqrt(1-x**2)
```

```python
def under_func(x, y):
    if y <= func(x):
        return True
    else:
        return False
```

```python
guess(100000, under_func)*4
```

```python
guess2(100000, func)*4
```

### 1. Метод проб и ошибок Монте-Карло

```python
x = list()
y = list()
for i in tqdm(range(1, 10000)):
    x.append(i)
    y.append(abs(guess(i, under_func)*4 - pi))
```

```python
plt.plot(np.log(x), y)
plt.plot(np.log(x), np.array(x)**-0.5)
plt.xlabel('number of experiments, log')
plt.ylabel('error')
plt.title(u'Зависимость погрешности от числа испытаний')
plt.show()
```

### 2. Метод Монте-Карло выборочного среднего

```python
x = list()
y = list()
for i in tqdm(range(1, 10000)):
    x.append(i)
    y.append(abs(guess2(i, func)*4 - pi))
```

```python
plt.plot(np.log(x), y)
plt.plot(np.log(x), np.array(x)**-0.5)
plt.xlabel('number of experiments, log')
plt.ylabel('error')
plt.title(u'Зависимость погрешности от числа испытаний')
plt.show()
```
